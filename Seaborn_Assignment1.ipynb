## Data Description

- Survival : 0 = No, 1 = Yes
- pclass(Ticket Class) : 1 = 1st, 2 = 2nd, 3 = 3rd
- Sex(Gender) : Male, Female
- Age : Age in years
- SibSp : Number of siblings/spouses abroad the titanic
- Parch : Number of parents/children abrod the titanic
- Ticket : Ticket Number
- Fare : Passenger fare
- Cabin : Cabin Number
- Embarked : Port of Embarkation, C = Cherbourg, Q = Queenstown, S = Southampton

# Import Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Q9: Draw barplot represents survived people based on sex

```python
sns.set_style("white") 
sns.barplot(x='sex', y='survived',data=df);
```

# Q10: Draw catplot represents survived people based on embarked

```python
se = sns.catplot(x='survived', kind='count', data=df, col='embarked')
se.set_xticklabels(['0', '1'])
se.set_axis_labels('survived', 'Count');
```

# Q11: Draw boxplot represents distribution of male and female based on age and pclass

```python
male_df = df[df['sex'] == 'male']
female_df = df[df['sex'] == 'female']

male_female_df = pd.concat([male_df, female_df])

plt.figure(figsize=(20, 8))

sns.boxplot(x='sex', y='age', hue='pclass', data=male_female_df)

plt.xlabel('sex');
```

# Q12: Draw heatmap represents correlations between sibsp, parch, age, fare, and survived columns

```python
col = ['sibsp', 'parch', 'age', 'fare', 'survived']
df_heat = df[col]

matrix = df_heat.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(matrix, annot=True, fmt='.2f');
```
